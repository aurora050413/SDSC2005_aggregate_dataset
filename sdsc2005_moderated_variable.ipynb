```python
import pandas as pd
import statsmodels.formula.api as smf


# 创建交互项
df["Housing_price_Unemployment_rate"] = df["Housing_price"] * df["Unemployment_rate"]

# 回归模型
formula = "Fertility_rate ~ Housing_price + Unemployment_rate + Housing_price_Unemployment_rate"  # 包含主效应 (Housing_price, Z) 和交互项 (Housing_price_Z)
model = smf.ols(formula=formula, data=df).fit()

# 提取回归系数
beta_0 = model.params["Intercept"]  # 截距
beta_1 = model.params["Housing_price"]          # 自变量 Housing_price 的主效应
beta_2 = model.params["Unemployment_rate"]          # 修饰变量 Z 的主效应
beta_3 = model.params["Housing_price_Unemployment_rate"]        # 交互效应

# 输出结果
print("回归系数：")
print(f"β0 (截距) = {beta_0}")
print(f"β1 (Housing_price 的主效应) = {beta_1}")
print(f"β2 (Z 的主效应) = {beta_2}")
print(f"β3 (交互效应) = {beta_3}")

# 输出模型摘要
print("\n模型摘要：")
print(model.summary())
```
